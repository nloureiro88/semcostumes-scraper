```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import chromedriver_binary
import csv
import time
```

```python
url = 'http://semcostumes.blogspot.com/?view=classic'
options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)
driver.get(url)
wait = WebDriverWait(driver, 15)
wait.until(ec.visibility_of_element_located((By.XPATH, "//div[@id='main']")))

item_count = 0
items = []

while item_count < 670:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    items = driver.find_elements_by_xpath("//li[@class='item']")
    item_count = len(items)
    print(f"> Found {item_count} results on the page...")
    time.sleep(0)
    
print(f"Found {len(items)} results on the page!")
```

```python
posts = []
for item in items:
    post_id = int(item.get_attribute('data-id'))
    link = item.find_element_by_css_selector(".article-header .ribbon").get_property('href')
    title = item.find_element_by_css_selector(".title").text
    text = item.find_element_by_css_selector(".entry-content").text
    text_clean = text.replace("\n", " ")
    timestamp = item.find_element_by_css_selector("abbr.published").get_property('title')
    date = timestamp[:10]
    
    posts.append({
        'post_id': post_id,
        'date': date,
        'title': title,
        'text': text,
        'text_clean': text_clean,
        'link': link,
        'timestamp': timestamp
    })
```

```python
with open('data/recipes.csv', 'w') as file:
    writer = csv.DictWriter(file, fieldnames=posts[0].keys())
    writer.writeheader()
    writer.writerows(posts)
```
